```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Yhills_July12_Analytics/main/titanic-training-data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
sns.relplot(x="SibSp",y="PassengerId",data=df,color = 'red')
```

```python
sns.relplot(x="SibSp",y="PassengerId",hue='Embarked',data=df,color = 'red')
```

```python
sns.relplot(y="SibSp",x='Embarked',data=df)
```

```python
sns.relplot(x="Pclass",y='Age',data=df)
```

```python
sns.relplot(x="Pclass",y='Age',hue='Sex',data=df)
```

```python
sns.catplot(x="SibSp",y='Survived', kind ='box',data=df)
```

```python
sns.catplot(x="SibSp",y='Survived', kind ='swarm',data=df)
```

```python
sns.catplot(x="SibSp",y='Survived', kind ='violin',data=df)
```

```python
sns.catplot(x="SibSp",y='Survived', kind ='violin',split="true",data=df)
```

```python
sns.catplot(x="SibSp",y='Survived', kind ='point',data=df)
```
